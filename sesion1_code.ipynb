```python
import pandas as pd
import numpy as np
import math

from sklearn.impute import KNNImputer
from sklearn.impute import SimpleImputer
from sklearn.neighbors import LocalOutlierFactor

import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns

from manny_library import multiple_plot

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('car details v4.csv')
data.info()
```

## Limpieza de datos

```python
# Borrando columnas que no se utilizarán

data.drop([ 'Model', 'Location', 'Color', 'Max Power', 'Max Torque'], axis='columns', inplace=True)
```

```python
# Se limpia la variable Engine para convertirla en una variable numérica

# Se borra todos los valores ' cc'
data['Engine'] = data['Engine'].str.replace(' cc', '')

# Se convierte a variable tipo float
data['Engine'] = data['Engine'].astype('float64')
data["Engine"].dtype
```

```python
# Verificación y eliminación de registros duplicados

# Verificación de registros duplicados

data.loc[data.duplicated()]
```

```python
# Eliminación de registros duplicados
data = data.drop_duplicates()

# Reiniciar el indice
data.reset_index(drop=True, inplace=True)
```

```python
# Convertir la columna 'Price' de rupias a dolares

tasa_de_cambio = 0.012 # Tasa de cambio de rupias a dólares

data['Price'] = data['Price'] * tasa_de_cambio
```

## Preparacion de datos

### Valores atipicos de la variable de salida

```python
# Calculo de valores atípicos

# Calculo de Q1 t Q3
Q1 = np.percentile(data['Price'], 25, interpolation = 'midpoint')
Q3 = np.percentile(data['Price'], 75, interpolation = 'midpoint')

# Cálculo del rango intercuartil
IQR = Q3 - Q1

# Cálculo de valor mínimo y máximo para los valores atípicos
VAInf = Q1 - 1.5*IQR
VASup = Q3 + 1.5*IQR

print(f'Valor atípico leve inferior:{VAInf}')
print(f'Valor atípico leve superior:{VASup}')
```

```python
# Se eliminan los valores atípicos
data = data.drop(data[data['Price'] > VASup].index)

# Reiniciar el indice
data.reset_index(drop=True, inplace=True)
```

```python
# Histograma y caja de bigotes del precio de los vehículos

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.title('Car Price Distribution Plot')
sns.distplot(data['Price'])

plt.subplot(1,2,2)
plt.title('Car Price Spread')
sns.boxplot(x=data['Price'], palette='nipy_spectral')

plt.show()
```

### Imputaciones

```python
# Identificación de valores nulos para las variables numéricas
for col in data[num_vars]:
    print (col, np.sum(pd.isnull(data[col])))
```

```python
# Registros donde varias características (Length, Width, Height y Seating Capacity) son vacías
data.loc[data['Length'].isna() & data['Width'].isna() & data['Height'].isna() & data['Seating Capacity'].isna()]
```

```python
# Se eliminan aquellas observaciones donde todas sus variables numéricas son vacías NaN
data = data.dropna(subset=['Length', 'Width', 'Height', 'Seating Capacity'])
```

```python
# Identificación de valores nulos para las variables numéricas
for col in data[num_vars]:
    print (col, np.sum(pd.isnull(data[col])))
```

```python
#Imputación de variables numéricas por medio de KNNImputer

# crear un objeto KNNImputer con k=5
imputer = KNNImputer(n_neighbors=5)

# imputar los valores faltantes en el dataframe
df_imputed = imputer.fit_transform(data[num_vars])

# convertir el arreglo numpy devuelto por el imputer en un dataframe
df_imputed = pd.DataFrame(df_imputed, columns=num_vars)

# imprimir el dataframe con los valores imputados
df_imputed.head()
```

```python
# Identificación de valores nulos para las variables categorica
for col in data[cat_vars]:
    print (col, np.sum(pd.isnull(data[col])))
```

```python
# Graficar la frecuencia de Drivetrain

plt.figure(figsize=(5, 3))
data['Drivetrain'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Frecuencia de Drivetrain')
plt.xlabel('Drivetrain')
plt.ylabel('Frecuencia')
plt.xticks(rotation=0)  # Rotar etiquetas en el eje x
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Imputación de variables categóricas por medio de SimpleImputer

# crear un objeto SimpleImputer con la estrategia de más frecuente
imputer = SimpleImputer(strategy='most_frequent')

# imputar los valores faltantes en el dataframe
df_imputed2 = imputer.fit_transform(data[cat_vars])

# convertir el arreglo numpy devuelto por el imputer en un dataframe
df_imputed2 = pd.DataFrame(df_imputed2, columns=cat_vars)

# imprimir el dataframe con los valores imputados
df_imputed2
```

```python
# Integración de los dos dataframes (categoricos y numéricos) después de la imputación
data = pd.merge(df_imputed, df_imputed2, left_index=True, right_index=True)
data.head()
```

```python
# verificación de la imputación: No existen valores nulos
for col in data.columns:
    print (col, np.sum(pd.isnull(data[col])))
```

### Valores atipicos para las variables de entrada

```python
# utilización de LocalOutlierFactor con 13 vecinos para identificar valores atípicos en las variables de entrada


# crear un objeto LocalOutlierFactor con n_neighbors=13
lof = LocalOutlierFactor(n_neighbors=13, contamination=0.1)

# identificar los valores atípicos por medio de lof
y_pred = lof.fit_predict(data[num_vars])

# Identificar los índices de los valores atípicos
outliers = y_pred == -1

print('Indices de valores atípicos: ', data[outliers].index.tolist())
```

```python
# Eliminar los valores atípicos del dataframe original
data = data.loc[~outliers]
```

### Transformaciones

```python
# Eliminación de categorias pequeñas para la variable Fuel Type

data['Fuel Type'].value_counts().sort_values(ascending=False)
```

```python
# Diccionario con los reemplazos a realizar
reemplazos = { "CNG": "Otro"
              ,"Electric": "Otro"
              , "LPG": "Otro"
              , "CNG + CNG": "Otro"
              , "Petrol + CNG": "Otro"
              , "Petrol + LPG": "Otro"}


# Se reemplazan los valores de acuerdo con el diccionario creado anteriormente
data['Fuel Type'] = data['Fuel Type'].map(reemplazos).fillna(data['Fuel Type'])

# Visualización de las nuevos valores agrupados para la categoria 'Fuel Type'
data['Fuel Type'].value_counts().sort_values(ascending=False)
```

```python
# Configuración de formato
pd.reset_option('display.max_rows')
```

```python
# Discretización (binning) de la variable 'Make'

# Cantidad de instancias por Make
data['Make'].value_counts().sort_values(ascending=False)
```

```python
# Se calcula la mediana por marca 'Make'

data.groupby('Make').Price.median().sort_values(ascending=False)
```

```python
# Gráfica de visualización de precio medio por marca 'Make'
plt.figure(figsize=(6, 6))

data.groupby('Make').Price.median().sort_values(ascending=True).plot.barh()
plt.title("Precio medio por marca")
plt.ylabel("Make")
plt.xlabel("Price")
plt.show()
```

```python
df = data.groupby('Make').Price.median().sort_values(ascending=False)
df = df.reset_index()
df.head(10)
```

```python
# Se crea una nueva variable que establece el un bin de acuerdo al precio medio por marca

# Configuración de formato de visualización
pd.set_option('display.float_format', lambda x: '%.4f' % x)

# Se define el numero de bins que se quieren crear
numBins = 5

# Se identifica el maximo valor
max_value = df["Price"].max()

# Se crea la nueva variable PriceRange dando el mayor valor al mayor precio medio
df['PriceRange'] = df['Price'].apply(lambda x: int(round(x / (max_value/numBins))))


df
```

```python
# Se adiciona la nueva variable al dataset original
data = pd.merge(data, df[['Make','PriceRange']], on='Make', how='left')

# Se elimina la variable Make
data = data.drop(['Make'], axis=1)

data.info()
```

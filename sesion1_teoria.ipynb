# CRISP-DM (Cross-Industry Standard Process for Data Mining): 
Es una metodología estándar y probada para guiar proyectos de ciencia de datos, que proporciona un enfoque estructurado y flexible para resolver problemas complejos usando análisis de datos. Este modelo describe seis fases: Comprensión del negocio, Comprensión de los datos, Preparación de los datos, Modelado, Evaluación y Despliegue, aunque las tareas dentro de estas fases se pueden realizar en diferentes órdenes. 
## Las fases de CRISP-DM
1. Comprensión del negocio: Se define el objetivo del proyecto y se traducen las necesidades del negocio en un problema de minería de datos. 
2. Comprensión de los datos: Se recopilan los datos iniciales, se exploran y se verifica su calidad. 
3. Preparación de los datos: Se limpian y transforman los datos, seleccionando y formateando los datos para el modelado. 
4. Modelado: Se aplican diversas técnicas de modelado para encontrar patrones en los datos. 
5. Evaluación: Se evalúa el modelo creado para determinar su efectividad en la consecución del objetivo del negocio. 
6. Despliegue: Se implementa la solución o el modelo en el entorno de negocio.

## Aprendizaje supervisado: Cada ejemplo de entrenamiento incluye la entrada y su salida esperada (etiqueta).

- Regresion - variable de salida numerica, ej: precio de la casa

- Clasificacion - variable de salida categorica, ej: se vende o no la casa. Modelos de clasificacion me arrojan la probabilidad, en este caso, la prob de que se venda la casa, prob de que no se venda la casa.

## Aprendizaje no supervisado
- El modelo no sabe cuál es la salida correcta, solo observa los datos de entrada.

- No tiene variable de salida, arroja clusteres o agrupaciones

## Modelo

- Parametrico: Construyen la funcion que aproxima los datos de entrenamiento a la variable objetivo con un numero fijo de parametros. Y = const + b1X1 + bnXn

- No parametrico: No presuponen una forma concreta en el modelo a generar, no es una funcion matematica. Ej: KNN

1. PREPARACION DE DATOS:

- Gestion de variables categoricas: convertir variables categoricas en numericas, existen dos metodos: Label Encoding y One Hot Encoding.

- Gestion de variables numericas: variables poseen magnitudes diferentes por lo que toca normalizar y estandarizar los datos mediante: MinMAxScaler, Normalizacion (normalizer) y estandarizador (se puede hacer por la media o por el rango intercuartilico (EstRobusto), este ultimo facilita la exclusion de datos atipicos. Cual es el mejor? Podemos probar con ambos y ver cual me da mejor resultado.

2. EVALUACION DE MODELOS

Modelamiento predictivo:

- Metodo de retencion: Es un método que consiste separar los datos en datos de entrenamiento (training) y prueba (test). Ideal dejar un 10% o 20% de datos de prueba (datos reales, lo mas parecido a la realidad).

 



- Metodo Validacion Cruzada: Los datos de muestra se dividen en K subconjuntos (k-folds). Creo diferente pliegues de la siguiente manera: Uno de los subconjuntos se utiliza como datos de prueba y el resto (K-1) como datos de entrenamiento. Finalmente se realiza la media aritmética de los resultados de cada iteración para obtener un único resultado.

Modelamiento categorico:

- Matriz de confusion ( clasificacion ): F1 Score Es especialmente útil cuando tienes un conjunto de datos desbalanceado.

- Curva ROC (clasificacion): se halla el area bajo la curva.
